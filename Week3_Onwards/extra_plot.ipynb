```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from collections import Counter

pd.set_option('display.max_columns', None,'display.max_row', None)
# pd.set_option('display.max_rows', 1000)

default_figsize = (15,5)
```

```python
df = pd.read_csv('unified_csv.csv')
df.info()
```

```python
num_org = len(df)
num_org
```

```python
org_by_cat_gp = df.iloc[:, 7:11].sum()
org_by_cat_gp
```

```python
fig, ax = plt.subplots()

np.divide(org_by_cat_gp,num_org).plot.bar(figsize=default_figsize)

ax.set_title('Number of investors against investment type')
plt.grid()
plt.show()
```
