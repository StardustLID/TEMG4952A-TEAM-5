```python
# use keyword to search for relevant degrees
bachelor_searchfor = ['Bachelor', 'BS', 'BSc', 'B.S', 'Bsc',
'BENG', 'BEng', 'B.Eng', 'Beng', 'BA', 'Ba', 'B.A.',
'BBA', 'B.B.A.', 'B.Tech']

master_searchfor = ['Master', 'MPHIL', 'MPhil', 'Mphil', 'M.Phil', 'M.phil',
'MS', 'MSc', 'M.Sc', 'Msc', 'MENG', 'MEng', 'M.Eng.', 'M.eng',
'MA', 'M.A', 'MBA', 'M.B.A', 'Mba', 'M.BA', 'M.Ba']

phd_searchfor = ['PHD', 'Phd', 'PhD', 'P.HD', 'P.Hd', 'P.hd', 'P.H.D']

# categories can be formed from one or more above elementary keywords
df_with_degree = df
df_with_degree.dropna(axis=0, how='any', subset=['degree_type'], inplace = True) # not yet finished

df_bachelor = df_with_degree
df_bachelor = df_bachelor[df_bachelor['degree_type'].str.contains('|'.join(bachelor_searchfor))]
df_master = df_with_degree
df_master = df_master[df_master['degree_type'].str.contains('|'.join(master_searchfor))]
df_phd = df_with_degree
df_phd = df_phd[df_phd['degree_type'].str.contains('|'.join(phd_searchfor))]
```
