```python
import pandas as pd
```

```python
df1 = pd.read_csv("organizations_processed_1.csv")
df2 = pd.read_csv("organizations_processed_2.csv", names = df1.columns)
df3 = pd.read_csv("organizations_processed_3.csv", names = df1.columns)

frames = [df1, df2, df3]

df = pd.concat(frames)
```

```python
df.info()
```

```python
df['num_funding_rounds'] = df['num_funding_rounds'].fillna(-1)
df['total_funding_usd'] = df['total_funding_usd'].fillna(-1)
df['total_funding'] = df['total_funding'].fillna(-1)

df['total_funding_currency_code'] = df['total_funding_currency_code'].fillna(-1)
df['last_funding_on'] = df['last_funding_on'].fillna(-1)
df['city'] = df['city'].fillna(-1)
df['primary_role'] = df['primary_role'].fillna(-1)
df['roles'] = df['roles'].fillna(-1)
df['num_exits'] = df['num_exits'].fillna(0)

df['country_code'] = df['country_code'].fillna(-1)
df['state_code'] = df['state_code'].fillna(-1)

df['category_groups_list'] = df['category_groups_list'].fillna("")
df['category_list'] = df['category_list'].fillna("")

df['founded_on'] = df['founded_on'].fillna(-1)
df['closed_on'] = df['closed_on'].fillna(-1)
df.info()
```

```python
df.dropna(axis=0, how='any',inplace = True)
df
```

```python
df = df.reset_index(drop = True)
df['country_code'].value_counts()
```

```python
def enconding(state):
    if state == "-1":
        value = -1
    elif state =="USA":
        value = 0
    elif state == "GBR":
        value = 1
    elif state == "IND":
        value = 2
    elif state == "CAN":
        value = 3
    elif state == "CHI":
        value = 4
    else :
        value = 5
    return value
df['country_code'] = df['country_code'].apply(lambda x:enconding(x))
df['state_code'].value_counts()

def enconding_sc(state):
    if state == "-1":
        value = -1
    elif state =="CA":
        value = 0
    elif state == "NY":
        value = 1
    elif state == "TX":
        value = 2
    elif state == "FL":
        value = 3
    elif state == "MD":
        value = 4
    elif state == "AZ":
        value = 5
    elif state == "MS":
        value = 6
    elif state == "NJ":
        value = 7
    elif state == "NM":
        value = 8
    elif state == "NC":
        value = 9
    elif state == "SC":
        value = 10
    elif state == "PA":
        value = 11
    elif state == "WA":
        value = 12
    else :
        value = 13
    return value

df['state_code'] = df['state_code'].apply(lambda x:enconding_sc(x))
df['roles'].value_counts()
```

```python
cleanup_nums = {"company": 0, "investor,company": 1, "company,investor": 1, "company,school": 2,\
                "school,company":2, "school,investor,company":3, "investor,company,school":3,\
                "company,investor,school":4, "investor,school":5, "school,investor":5, "investor":6}
df = df.replace(cleanup_nums)
df['status'].value_counts()
```

```python
cleanup_nums = {"ipo": 0, "operating": 1, "acquired": 2, "closed": 3}
df = df.replace(cleanup_nums)
df['total_funding_currency_code'].value_counts()
```

```python
def enconding_cur(state):
    if state == "-1":
        value = -1
    elif state =="USD":
        value = 0
    elif state == "GBP":
        value = 1
    elif state == "EUR":
        value = 2
    elif state == "CNY":
        value = 3
    elif state == "JPY":
        value = 4
    elif state == "CHF":
        value = 5
    elif state == "AUD":
        value = 6
    elif state == "NZD":
        value = 7
    elif state == "MXN":
        value = 8
    elif state == "SGD":
        value = 9
    elif state == "SGD":
        value = 10
    else :
        value = 11
    return value
df['total_funding_currency_code'] = df['total_funding_currency_code'].apply(lambda x:enconding_cur(x))
df
```
